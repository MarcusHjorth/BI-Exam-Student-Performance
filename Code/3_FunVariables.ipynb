```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn tools
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Set seaborn style for clean visuals
sns.set(style="whitegrid")

# Load dataset
student = pd.read_csv('../Data/student_habits_performance.csv')

# Clean dataset (there are no duplicates or null values, so no action needed for deleting those)
print(student.isnull().sum())
print("Duplicate values: ", student.duplicated().sum())

student.columns
```

```python
correlations = student.corr(numeric_only=True)

plt.figure(figsize=(12, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title(" Korrelation mellem faktorer og eksamensscore")
plt.show()
```

```python
# Studty hours
sns.regplot(x='study_hours_per_day', y='exam_score', data=student)
plt.title(" Study Hours vs Exam Score")
plt.show()

# Social media
sns.regplot(x='social_media_hours', y='exam_score', data=student)
plt.title(" Social Media Usage vs Exam Score")
plt.show()

# Sleep
sns.regplot(x='sleep_hours', y='exam_score', data=student)
plt.title(" Sleep Hours vs Exam Score")
plt.show()


# 
sns.regplot(x='attendance_percentage', y='exam_score', data=student)
plt.title("Attendance_percentage vs Exam Score")
plt.show()
```
